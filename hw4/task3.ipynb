```python
"""
In Jupyter, create a bar chart showing the number of occurrences of the overall most abundant complaint type over the 
 first 2 months of 2020. (note: before making the plot, you'll need to figure out the complaint type to use)
How does this compare to that same complaint type in June-July of 2020?
"""

import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('trimmed_data.csv', low_memory=False, header=None, on_bad_lines='skip', sep=',')
```

```python
df[2] = pd.to_datetime(df[2], format='%Y-%m-%d %H:%M:%S', errors='coerce')
jan_feb = df[((df[2].dt.month == 1) | (df[2].dt.month == 2))]
complaint_counts_jan_feb = jan_feb[5].value_counts()
most_common_complaint_jan_feb = complaint_counts_jan_feb.idxmax()

```

```python
jun_jul = df[((df[2].dt.month == 6) | (df[2].dt.month == 7))]
complaint_counts_jun_jul = jun_jul[5].value_counts()
most_common_complaint_jun_jul = complaint_counts_jun_jul.idxmax()

```

```python
labels = ['Jan-Feb 2020', 'Jun-Jul 2020']
values = [complaint_counts_jan_feb[most_common_complaint_jan_feb], complaint_counts_jun_jul[most_common_complaint_jun_jul]]

plt.bar(labels, values)
plt.ylabel('Number of Occurrences')
plt.xlabel('Time Period')
plt.title(f'Most Common Complaint Type: {most_common_complaint_jan_feb}')
plt.show()
```
